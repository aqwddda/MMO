Загрузка модели для русского языка и подготовка текста для анализа

```python
nlp = spacy.load("ru_core_news_sm")
text = "Мне нравится анализировать текст с помощью spaCy."
```

Токенизация текста

```python
doc = nlp(text)
tokens = [token.text for token in doc]
print("Токены:", tokens)
```

Частеречная разметка

```python
pos_tags = [(token.text, token.pos_) for token in doc]
print("Частеречная разметка:", pos_tags)
```

Лемматизация

```python
lemmas = [token.lemma_ for token in doc]
print("Леммы:", lemmas)
```

Выделение именованных сущностей

```python
entities = [(entity.text, entity.label_) for entity in doc.ents]
print("Именованные сущности:", entities)
```

Разбор предложения

```python
parse_info = [(token.text, token.dep_) for token in doc]
print("Разбор предложения:", parse_info)
```

Визуализация токенизации и разбора предложения

```python
from spacy import displacy

# Визуализация токенизации
displacy.render(doc, style="dep", jupyter=True)

# Визуализация именованных сущностей
displacy.render(doc, style="ent", jupyter=True)
```
