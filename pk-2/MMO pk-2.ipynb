 Импорт необходимых библиотек и загрузка данных

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

Загрузка данных

```python
# Загрузка набора данных IMDb
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)
```

Подготовка данных

```python
# Импортируем инструмент для преобразования индексов в слова
word_index = imdb.get_word_index()

# Переворачиваем word_index, чтобы можно было вызывать слова по их числовым индексам
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

# Функция для декодирования отзывов
def decode_review(text):
    return ' '.join([reverse_word_index.get(i, '?') for i in text])

# Декодируем отзывы
X_train = [decode_review(text) for text in train_data]
X_test = [decode_review(text) for text in test_data]
y_train = train_labels
y_test = test_labels

```

Векторизация признаков

```python
# Векторизация признаков
count_vect = CountVectorizer(max_features=5000)
X_train_counts = count_vect.fit_transform(X_train)
X_test_counts = count_vect.transform(X_test)

tfidf_vect = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf_vect.fit_transform(X_train)
X_test_tfidf = tfidf_vect.transform(X_test)
```

Обучение и оценка моделей

```python
# Обучение и оценка моделей
logreg = LogisticRegression(max_iter=1000, solver='liblinear')

logreg.fit(X_train_counts, y_train)
y_pred_logreg_counts = logreg.predict(X_test_counts)
print("LogisticRegression + CountVectorizer:")
print(classification_report(y_test, y_pred_logreg_counts))

logreg.fit(X_train_tfidf, y_train)
y_pred_logreg_tfidf = logreg.predict(X_test_tfidf)
print("LogisticRegression + TfidfVectorizer:")
print(classification_report(y_test, y_pred_logreg_tfidf))
```
